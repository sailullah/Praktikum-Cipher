```python
def caesar_encrypt(text, shift):
    result = ''
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

def caesar_decrypt(cipher_text, shift):

    return caesar_encrypt(cipher_text, -shift)


text = input("Masukkan teks yang ingin dienkripsi: ")
shift = int(input("Masukkan jumlah pergeseran (shift): "))


encrypted = caesar_encrypt(text, shift)
decrypted = caesar_decrypt(encrypted, shift)


print("\n=== HASIL ===")
print(f"Teks Asli     : {text}")
print(f"Hasil Enkripsi: {encrypted}")
print(f"Hasil Dekripsi: {decrypted}")


with open("hasil_cipher.txt", "w") as file:
    file.write("=== HASIL ENKRIPSI DAN DEKRIPSI ===\n")
    file.write(f"Teks Asli     : {text}\n")
    file.write(f"Hasil Enkripsi: {encrypted}\n")
    file.write(f"Hasil Dekripsi: {decrypted}\n")

print("\nHasil sudah disimpan ke file 'hasil_cipher.txt'")
```

```python
def vigenere_encrypt(plain, key):
    key = key.upper()
    result = ''
    for i, char in enumerate(plain.upper()):
        if char.isalpha():
            shift = ord(key[i % len(key)]) - 65
            result += chr((ord(char) - 65 + shift) % 26 + 65)
        else:
            result += char
    return result

def vigenere_decrypt(cipher, key):
    key = key.upper()
    result = ''
    for i, char in enumerate(cipher.upper()):
        if char.isalpha():
            shift = ord(key[i % len(key)]) - 65
            result += chr((ord(char) - 65 - shift) % 26 + 65)
        else:
            result += char
    return result


text = input("Masukkan teks yang ingin dienkripsi: ")
key = input("Masukkan kunci (key): ")


encrypted = vigenere_encrypt(text, key)
decrypted = vigenere_decrypt(encrypted, key)


print("\n=== HASIL ===")
print(f"Teks Asli     : {text}")
print(f"Hasil Enkripsi: {encrypted}")
print(f"Hasil Dekripsi: {decrypted}")


with open("hasil_vigenere.txt", "w") as file:
    file.write("=== HASIL ENKRIPSI DAN DEKRIPSI (Vigenere Cipher) ===\n")
    file.write(f"Teks Asli     : {text}\n")
    file.write(f"Kunci          : {key}\n")
    file.write(f"Hasil Enkripsi : {encrypted}\n")
    file.write(f"Hasil Dekripsi : {decrypted}\n")

print("\nHasil sudah disimpan ke file 'hasil_vigenere.txt'")
```

```python
def egcd(a, b):
    """Extended Euclidean Algorithm, mengembalikan (g, x, y) sehingga ax + by = g = gcd(a,b)."""
    if b == 0:
        return (a, 1, 0)
    g, x1, y1 = egcd(b, a % b)
    return (g, y1, x1 - (a // b) * y1)

def modinv(a, m):
    """Kembalikan invers modular a (mod m). Jika tidak ada, raise ValueError."""
    g, x, _ = egcd(a, m)
    if g != 1:
        raise ValueError(f"'a' tidak invertible modulo {m} (gcd({a},{m}) = {g}). Pilih a yang relatif prima dengan {m}.")
    return x % m

def affine_encrypt(text, a, b):
    """Enkripsi Affine, mempertahankan case."""
    result = ''
    for char in text:
        if char.isalpha():
            if char.isupper():
                base = ord('A')
            else:
                base = ord('a')
            x = ord(char) - base
            # E(x) = (a*x + b) mod 26
            result += chr(((a * x + b) % 26) + base)
        else:
            result += char
    return result

def affine_decrypt(cipher_text, a, b):
    """Dekripsi Affine menggunakan invers modular a."""
    a_inv = modinv(a, 26)
    result = ''
    for char in cipher_text:
        if char.isalpha():
            if char.isupper():
                base = ord('A')
            else:
                base = ord('a')
            y = ord(char) - base
            # D(y) = a_inv * (y - b) mod 26
            result += chr((a_inv * (y - b) % 26) + base)
        else:
            result += char
    return result


try:
    text = input("Masukkan teks: ")
    a = int(input("Masukkan nilai a (harus relatif prima dengan 26, mis. 1,3,5,7,11,15,17,19,21,23,25): "))
    b = int(input("Masukkan nilai b (0-25): "))

    encrypted = affine_encrypt(text, a, b)
    decrypted = affine_decrypt(encrypted, a, b)

    print("\n=== HASIL ===")
    print(f"Teks Asli     : {text}")
    print(f"Hasil Enkripsi: {encrypted}")
    print(f"Hasil Dekripsi: {decrypted}")


    with open("hasil_affine.txt", "w") as file:
        file.write("=== HASIL ENKRIPSI DAN DEKRIPSI (Affine Cipher) ===\n")
        file.write(f"Teks Asli      : {text}\n")
        file.write(f"Nilai a        : {a}\n")
        file.write(f"Nilai b        : {b}\n")
        file.write(f"Hasil Enkripsi : {encrypted}\n")
        file.write(f"Hasil Dekripsi : {decrypted}\n")

    print("\nHasil sudah disimpan ke file 'hasil_affine.txt'")

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Terjadi kesalahan:", e)
```

```python
def generate_table(key):
    """Membuat tabel 5x5 berdasarkan key."""
    alphabet = 'ABCDEFGHIKLMNOPQRSTUVWXYZ'
    table = ''
    for c in key.upper() + alphabet:
        if c not in table:
            table += c
    return [table[i:i+5] for i in range(0, 25, 5)]

def find_position(table, char):
    """Mencari posisi huruf dalam tabel."""
    for i, row in enumerate(table):
        if char in row:
            return i, row.index(char)
    return None

def playfair_encrypt(plaintext, key):
    """Enkripsi Playfair Cipher."""
    table = generate_table(key)
    plaintext = plaintext.upper().replace('J', 'I')
    plaintext = ''.join([c for c in plaintext if c.isalpha()])


    pairs = []
    i = 0
    while i < len(plaintext):
        a = plaintext[i]
        b = plaintext[i + 1] if i + 1 < len(plaintext) else 'X'
        if a == b:
            pairs.append(a + 'X')
            i += 1
        else:
            pairs.append(a + b)
            i += 2


    result = ''
    for a, b in pairs:
        r1, c1 = find_position(table, a)
        r2, c2 = find_position(table, b)
        if r1 == r2:
            result += table[r1][(c1 + 1) % 5]
            result += table[r2][(c2 + 1) % 5]
        elif c1 == c2:
            result += table[(r1 + 1) % 5][c1]
            result += table[(r2 + 1) % 5][c2]
        else:
            result += table[r1][c2]
            result += table[r2][c1]
    return result

def playfair_decrypt(ciphertext, key):
    """Dekripsi Playfair Cipher."""
    table = generate_table(key)
    ciphertext = ciphertext.upper().replace('J', 'I')
    ciphertext = ''.join([c for c in ciphertext if c.isalpha()])

    result = ''
    for i in range(0, len(ciphertext), 2):
        a = ciphertext[i]
        b = ciphertext[i + 1]
        r1, c1 = find_position(table, a)
        r2, c2 = find_position(table, b)
        if r1 == r2:
            result += table[r1][(c1 - 1) % 5]
            result += table[r2][(c2 - 1) % 5]
        elif c1 == c2:
            result += table[(r1 - 1) % 5][c1]
            result += table[(r2 - 1) % 5][c2]
        else:
            result += table[r1][c2]
            result += table[r2][c1]
    return result


text = input("Masukkan teks yang ingin dienkripsi: ")
key = input("Masukkan kunci (key): ")


encrypted = playfair_encrypt(text, key)
decrypted = playfair_decrypt(encrypted, key)


print("\n=== HASIL ===")
print(f"Teks Asli     : {text}")
print(f"Kunci          : {key}")
print(f"Hasil Enkripsi : {encrypted}")
print(f"Hasil Dekripsi : {decrypted}")


with open("hasil_playfair.txt", "w") as file:
    file.write("=== HASIL ENKRIPSI DAN DEKRIPSI (Playfair Cipher) ===\n")
    file.write(f"Teks Asli      : {text}\n")
    file.write(f"Kunci           : {key}\n")
    file.write(f"Hasil Enkripsi  : {encrypted}\n")
    file.write(f"Hasil Dekripsi  : {decrypted}\n")

print("\nHasil sudah disimpan ke file 'hasil_playfair.txt'")
```

```python
import numpy as np

def mod_inverse_matrix(matrix, mod):
    """Mencari invers matriks dalam modulo tertentu (jika ada)."""
    det = int(round(np.linalg.det(matrix)))
    det_inv = None
    det = det % mod

    for i in range(mod):
        if (det * i) % mod == 1:
            det_inv = i
            break
    if det_inv is None:
        raise ValueError("Determinannya tidak memiliki invers modulo 26, pilih kunci lain.")


    matrix_mod_inv = (
        det_inv * np.round(det * np.linalg.inv(matrix)).astype(int)
    ) % mod
    return matrix_mod_inv.astype(int)

def hill_encrypt(text, key):
    """Enkripsi Hill Cipher."""
    text = text.upper().replace(" ", "")
    n = int(len(key) ** 0.5)
    key_matrix = np.array(key).reshape(n, n)
    result = ''


    if len(text) % n != 0:
        text += 'X' * (n - len(text) % n)

    for i in range(0, len(text), n):
        block = [ord(c) - 65 for c in text[i:i+n]]
        cipher = np.dot(key_matrix, block) % 26
        result += ''.join(chr(c + 65) for c in cipher)
    return result

def hill_decrypt(cipher, key):
    """Dekripsi Hill Cipher."""
    cipher = cipher.upper().replace(" ", "")
    n = int(len(key) ** 0.5)
    key_matrix = np.array(key).reshape(n, n)
    key_inv = mod_inverse_matrix(key_matrix, 26)
    result = ''

    for i in range(0, len(cipher), n):
        block = [ord(c) - 65 for c in cipher[i:i+n]]
        plain = np.dot(key_inv, block) % 26
        result += ''.join(chr(int(c) + 65) for c in plain)
    return result


try:
    text = input("Masukkan teks yang ingin dienkripsi: ").strip()
    key_input = input("Masukkan kunci matriks (mis. 3 3 2 5 untuk matriks 2x2): ").split()
    key = list(map(int, key_input))

    encrypted = hill_encrypt(text, key)
    decrypted = hill_decrypt(encrypted, key)

    print("\n=== HASIL ===")
    print(f"Teks Asli     : {text}")
    print(f"Kunci Matriks : {key}")
    print(f"Hasil Enkripsi: {encrypted}")
    print(f"Hasil Dekripsi: {decrypted}")


    with open("hasil_hill.txt", "w") as file:
        file.write("=== HASIL ENKRIPSI DAN DEKRIPSI (Hill Cipher) ===\n")
        file.write(f"Teks Asli      : {text}\n")
        file.write(f"Kunci Matriks  : {key}\n")
        file.write(f"Hasil Enkripsi : {encrypted}\n")
        file.write(f"Hasil Dekripsi : {decrypted}\n")

    print("\nHasil sudah disimpan ke file 'hasil_hill.txt'")

except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("Terjadi kesalahan:", e)
```
